```python
#42
A = np.random.randint(0, 10, 10)
B = np.random.randint(0, 10, 10)
print(np.equal(A, B))
print(np.allclose(A, B))
print(np.array_equal(A, B))
```

```python
#43
A = np.random.random(10)
A[0] = 1.11
A.flags.writeable = False
try:
    A[0] = 1.22
except Exception:
    print("Cannot write to array")
```

```python
#44
cartesian = np.random.randint(1, 10, (10, 2))
cartesian = cartesian.astype(np.float32)
print(cartesian)
#using apply along axis
get_radius = lambda x: np.sqrt(np.sum(np.square(x)))
get_theta = lambda x: np.arctan2(x[1], x[0])
get_polar = lambda x:  np.array([get_radius(x), get_theta(x)])
print(np.apply_along_axis(get_polar, -1, cartesian))


#using indexing
X, Y = cartesian[:, 0], cartesian[:, 1]
R = np.sqrt(X**2 + Y**2)
T = np.arctan2(Y,X)
print(R)
print(T)
```

```python
#45
Z = np.random.random(10)
print(Z)
Z[Z.argmax()] = 0
print(Z)
```

```python
#46
#using np.meshgrid
Z = np.zeros((5,5), [('x', np.float32), ('y', np.float32)])
print(Z)
Z['x'], Z['y'] = np.meshgrid(np.linspace(0, 1, 5), np.linspace(1, 0, 5))
print(Z)
```

```python
#47
X = np.arange(8)
Y = X + 0.5
C = 1.0 / np.subtract.outer(X, Y)
print(C)
print(np.linalg.det(C))
```

```python
for dtype in [np.int8, np.int32, np.int64]:
   print(np.iinfo(dtype).min)
   print(np.iinfo(dtype).max)
for dtype in [np.float32, np.float64]:
   print(np.finfo(dtype).min)
   print(np.finfo(dtype).max)
   print(np.finfo(dtype).eps)
```

```python
#50
A = np.random.randint(1, 100, (4,))
print(A)
Z = np.random.uniform(0, 100, (4,))
print(Z)
index = (np.abs(A-Z)).argmin()
print(index)
print(Z[index])
```
